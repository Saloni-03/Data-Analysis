```python
df.drop(['S. No.'],axis=1,inplace=True)
```

```python
df['total_cases'] = df['Total Confirmed cases (Indian National)']+df['Total Confirmed cases ( Foreign National )']
```

```python
total_cases_overall=df['total_cases'].sum()
print("Total cases in India are",total_cases_overall)
```

```python
df['active_cases'] = df['Total Confirmed cases (Indian National)']-df['Cured']-df['Death']
```

```python
df_world = pd.read_csv('https://raw.githubusercontent.com/mona2401/COVID-19-Data-Analyisis/master/covid_19_data.csv')
```

```python
df_world
```

```python
df_world.rename(columns={'ObservationDate':'Date','Country/Region':'Country'},inplace = True)
```

```python
df_world
```

```python
confirmed = df_world.groupby('Date').sum()['Confirmed'].reset_index()
death = df_world.groupby('Date').sum()['Deaths'].reset_index()
recovered = df_world.groupby('Date').sum()['Recovered'].reset_index()
```

```python
confirmed
death
recovered
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'],y=confirmed['Confirmed'],mode = 'lines+markers',name = 'Confirmed',line = dict(color = 'blue',width = 2)))
fig.add_trace(go.Scatter(x=death['Date'],y=death['Deaths'],mode = 'lines+markers',name = 'Deaths',line = dict(color = 'red',width = 2)))
fig.add_trace(go.Scatter(x=recovered['Date'],y=recovered['Recovered'],mode = 'lines+markers',name = 'Recovered',line = dict(color = 'green',width = 2)))
```
